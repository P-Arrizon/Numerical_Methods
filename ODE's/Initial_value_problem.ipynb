Our goal with differential functions is to find a function $f(x)$ that can satisfy the properties we specify. Sometimes those properties can be initial values that we can use to solve our ODE. 

##  1.) **Euler's Method**

The methodology behind Euler's method is that we re-evaluate our slope or change after a certain step size, generally a smaller step size will allow us to approximate the function more accurate. 

##### **EULERS METHOD**

(6.7)

$ w_0 = y_0 \\~\\ w_{i+1} = w_i + hf(t_i, w_i) $

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline

def f(t,w):
    '''
    We want to first define our function that we want to approximate, this function should be equal to y' 
    '''
    wi = t*w + t**3
    return wi

def ftrue(t):
    '''
    define our true function if you want to plot the error
    '''
    y = 3*np.e ** (t ** 2 / 2) - t ** 2 - 2

    return y


def euler(w0,h,a,b):
    '''
    This is straight from (6.7)

    params: (initial value (w0), step size (h), intervals[a,b])
    '''

    ti = np.arange(a,b+h,h)
    num_of_int = len(ti)
    wi = np.zeros(num_of_int)
    wi[0] = w0

    for n in range(num_of_int-1):
        wi[n+1] = wi[n] + h * f(ti[n], wi[n])

    x_test = np.linspace(a,b,1000)
    y_true = [ftrue(i) for i in x_test]

    plt.plot(x_test,y_true,'b-',ti,wi,'r.-')
    plt.legend(['True','Euler approx'])
    plt.grid()
    plt.title("Euler vs True")
    plt.show()




```

```python
euler(1,0.2,0,1)
```
