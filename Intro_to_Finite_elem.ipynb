## Reasons why we would use Finite Element analysis

1. Most real world problems involve complicated features like shape and loads. These non linear attributes do not allow us to use any type of analytical solution forcing us to use numerical methods like FEA (Finite element analysis). 

2. We can use computers to help us better understand complicated systems better than what we could if we analyzed it as a whole. We can break down the structure into smaller pieces that allow deeper analysis. By doing this we better utilize our time and recourses. 



# Pendulum Problem 

We know that a pendulum is governed by the first order differential equations below:

$ \frac{du}{dt} = f(t,u), $ $ \frac{d\theta}{dt} = v,\ \frac{dv}{dv} = -\lambda ^ 2\theta $

We can attempt to solve this differential equation by approximating the derivate using the following:

$ (\frac{du}{dt})|_{t = ti} \approx \frac{u(t_{i+1}) - u(t_i)}{t_{i+1} - t_i}$

The equation is an approximation due to the fact that we did not take the limit as $\Delta t$ approaches 0. We use a computer to find the right *step value* that gives us an answer within the tolerance we need. 

```python
def circle_check(x,y):
    L = .1
    h = .05
    r = 0.02 # lets assume the middle of the circle is in the middle of the block

    d = ((x)** 2 + (y) ** 2) ** 0.5

    if d  >  r :
        return 1
    else :
        return 

    
    
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.spatial import Delaunay


L = .1
h = .05
r = 0.02 # lets assume the middle of the circle is in the middle of the block
n_elements = 25
nodes = []
p = []
counter = 0


for x in np.linspace(0,2*L, int(n_elements)):
    for y in np.linspace(0,2*h, int(n_elements)): 
        if circle_check(x,y) == 1:
            nodes.append([x,y])
            counter += 1
        else:
            p.append([x,y])
        

points = np.array(nodes)
tri = Delaunay(points)
plt.triplot(points[:,0], points[:,1], tri.simplices)
plt.plot(points[:,0], points[:,1], 'o')

plt.show()







    
```
